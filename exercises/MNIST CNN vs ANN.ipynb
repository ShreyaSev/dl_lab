```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout,Conv2D,MaxPool2D

from keras.utils import to_categorical

from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
x_train.shape
```

```python
x_train_Ann = x_train.reshape(len(x_train),784)
x_test_Ann = x_test.reshape(len(x_test),784)
```

```python
x_train_Ann.shape
```

```python
ANN = Sequential()
ANN.add(Dense(128,input_shape=(784,)))
ANN.add(Dense(64,activation='relu'))
ANN.add(Dense(32,activation='relu'))
ANN.add(Dense(10,activation='softmax'))
```

```python
ANN.summary()
```

```python
ANN.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history_ANN = ANN.fit(x_train_Ann,y_train,batch_size=128,epochs=10,validation_data=(x_test_Ann,y_test))
```

```python
ANN.evaluate(x_test_Ann,y_test)
```

```python
x_train_Cnn = x_train.reshape(len(x_train),28,28,1)
x_test_Cnn = x_test.reshape(len(x_test),28,28,1)
```

```python
x_train_Cnn.shape
```

```python
CNN = Sequential()
CNN.add(Conv2D(64,(3,3),input_shape=(28,28,1),padding='same'))
CNN.add(MaxPool2D(2,2))
CNN.add(Conv2D(32,(3,3),padding='same',activation='relu'))
CNN.add(MaxPool2D(2,2))
CNN.add(Flatten())
CNN.add(Dense(64,activation='relu'))
CNN.add(Dense(10,activation='softmax'))
```

```python
CNN.summary()
```

```python
CNN.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history_Cnn = CNN.fit(x_train_Cnn,y_train,epochs=10,batch_size=32,validation_data=(x_test_Cnn,y_test))
```

```python
CNN.evaluate(x_test_Cnn,y_test)
```
